```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dat=pd.read_csv('/content/drive/MyDrive/India_Menu.csv')
```

```python
dat.dtypes
```

```python
data=dat.drop(['Menu Category','Menu Items','Per Serve Size'],axis=1)
data
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
```

```python
y=data['Energy (kCal)']
x=data.drop(['Energy (kCal)'],axis=1)
```

```python
# Split into testing and training
# testing and training library avalable in the sklearn library
# sklearn-ppt
from sklearn.model_selection import train_test_split
# random state is a counter, which decides in which pattern the dataset should 
# be taken for testing and training
# test size is 0.25 that is 25% data for testing and 75% data for training
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
x_test
```

```python
y_test
```

```python
# make logistic regression model
from sklearn import linear_model
# create instance of the model
lr=linear_model.LinearRegression()
# now fitting the instances created to the training data and creating the 
# linear regression model and store int another variable
model=lr.fit(x_train,y_train)
```

```python
# predict the output
predictions=model.predict(x_test)
predictions
```

```python
print(predictions)
```

```python
# by comparing these values mean squared error is found .for that we import mean squared error from sklearn
```

```python
from sklearn.metrics import mean_squared_error
print('MSC is:',mean_squared_error(y_test,predictions))
```
