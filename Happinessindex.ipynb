```python
q1=np.percentile(data['Generosity'],25,interpolation='midpoint')
q2=np.percentile(data['Generosity'],50,interpolation='midpoint')
q3=np.percentile(data['Generosity'],75,interpolation='midpoint')
print(q1)
print(q2)
print(q3)
```

```python
iqr=q3-q1

low_limit=q1-1.5*iqr
up_limit=q3-+1.5*iqr
print(low_limit)
print(up_limit)
```

```python
outlier=[]
for x in data['Generosity']:
  if(x<low_limit) and (x>up_limit):
    outlier.append(x)
outlier
```
